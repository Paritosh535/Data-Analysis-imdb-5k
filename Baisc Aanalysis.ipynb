### Data Preprocess Required for Analysis

#### Most variables are left-skewed. So that here we use median/mode for  fill missing values.

```python
columns_median=['num_critic_for_reviews','director_facebook_likes','actor_3_facebook_likes','actor_1_facebook_likes','num_user_for_reviews','gross','budget','actor_2_facebook_likes','movie_facebook_likes']
columns_mode=['color','duration','facenumber_in_poster','language','country','content_rating','title_year','aspect_ratio']
median=df[columns_median].median()
mode=df[columns_mode].mode().iloc[0]
df=df.fillna(median)
df=df.fillna(mode)
```

#### Need to Remove constant columns (unique)
#### Need to  remove duplicate columns (drop_duplicates)
#### Need to  remove duplicate rows (drop_duplicates)

```python
duplicates=df.duplicated(subset='movie_title', keep=False)
sum(duplicates) ## Number of duplicates based on movie title
```

#### drop duplicate rows

```python
print("Total Rows {} dupicate rows {}".format(len(df),len(df)-len(df.drop_duplicates())))
df=df.drop_duplicates()
print("Total Remain After Delete Duplicate rows {}".format(len(df)))
```

### Add columns 
       1) Profit_Loss columns with help of gross-budget
       2) Movies_type base on imdb_score rating (low,avg,medium,high

```python
df["profit_loss"]=((df["gross"]-df["budget"]))
```

```python
def movie_type(rating):
    if rating < 5:
        return "Low"
    elif rating == 5:
        return "Avarage"
    elif rating > 5 and rating < 8:
        return "Medium"
    elif rating >=8 and rating <= 10:
        return "High"
    
df["movies_type"]=df["imdb_score"].apply(movie_type)
```

#### Drop rows contain NaN value

```python
df= df.dropna()
```

#### Assign Dtype to this Dataset. DataFrame Optimization memory usage 1.1 MB 

```python
df["title_year"] = df["title_year"].astype("int")
colObj = list(df.select_dtypes(include="O").columns)
colNonObject = list(df.select_dtypes(exclude="O").columns)
colObj.append("title_year") ## default it is float but actual it is categorical vairable 

# don't have any boolean feature
for col in colObj:
    df[col]=df[col].astype("category")
    
#convert to datetime dtype
# df['title_year'] = pd.to_datetime(df['title_year'],format="%Y")

# Make int rest of the feature except aspect_ratio","title_year","imdb_score"
for col in set(colNonObject) - set(["aspect_ratio","title_year","imdb_score"]):
    df[col] = df[col].astype("int")
```
